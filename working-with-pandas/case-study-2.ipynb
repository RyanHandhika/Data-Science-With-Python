```python
import pandas as pd
import numpy as np

order_history = pd.DataFrame({
  "order_id": [1, 2, 3, 4, 5, 6, 7, 8],
  "cust_no": ["C1", "C2", "C3", "C5", "C6", "C7", "C8", "C10"],
  "order_amount": [100, 200, 250, 150, 300, 350, 100, 150]
})

customer_base = pd.DataFrame({
  "cust_id": ["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10"],
  "city": ["Bandung", "Jakarta", "Tangerang", "Tangerang", "Jakarta", "Jakarta", "Jakarta", "Bandung", "Tangerang", "Jakarta"],
})
```

```python
order_history
```

```python
customer_base
```

```python
df_merge = order_history.merge(customer_base, how="left", left_on="cust_no", right_on="cust_id")
```

```python
df_merge
```

```python
result = pd.DataFrame(df_merge.groupby(["city"])["order_amount"].sum())
```

```python
result
```

```python
result.sort_values("order_amount", ascending=False)
```

```python
df_new_data =order_history.merge(customer_base, how="inner", left_on="cust_no", right_on="cust_id").drop(columns=["cust_id"])
```

```python
df_new_data
```

```python
df_new_data.groupby(["city"]).agg({"order_amount": "sum"})
```

```python
# harus pake reset index agar city dengan order_amount itu berada di level yang sama
data_sum_order = df_new_data.groupby(["city"]).agg({"order_amount": "sum"}).reset_index()
```

```python
data_sum_order
```

```python
data_sum_order.columns = ["city", "total_order_amount"]
```

```python
data_sum_order
```

```python
data_sum_order.sort_values("total_order_amount", ascending = False).reset_index(drop = True)
```
