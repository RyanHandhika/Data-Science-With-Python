```python
import pandas as pd
import numpy as np

customers = {
  "id": [1, 2, 3, 4],
  "name": ["Alice", "Bob", "Charlie", "David"]
}

orders = {
  "id": [1, 2],
  "customer_id": [1, 3]
}

df_customers = pd.DataFrame(customers)
df_orders = pd.DataFrame(orders)
```

```python
# Gemini
df_inactive = df_customers.loc[~df_customers["id"].isin(df_orders["customer_id"]), ["name"]]

# Me
# df_inactive = df_customers[~(df_customers["id"].isin(df_orders["customer_id"]))]
# df_inactive.loc[:, ["name"]]
```
