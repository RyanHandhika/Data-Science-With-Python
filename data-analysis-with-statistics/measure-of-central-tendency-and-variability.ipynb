```python
import statistics as stats
import numpy as np

data_num = np.random.randint(0, 100, 100)
data_cat = ["S", "M", "L", "M", "XL", "XXL", "S"]
```

```python
data_num
```

```python
data_cat
```

```python
# Central Tendency

stats.mean(data_num)
```

```python
stats.median(data_num)
```

```python
stats.mode(data_num)
```

```python
stats.mode(data_cat)
```

```python
stats.multimode(data_cat)
```

```python
# Variability

np.var(data_num)
```

```python
np.std(data_num)
```

```python
np.max(data_num) - np.min(data_num)
```

```python
print(f'Q1: {np.quantile(data_num, 0.25)}')
print(f'Q2: {np.quantile(data_num, 0.5)}')
print(f'Q3: {np.quantile(data_num, 0.75)}')
```
