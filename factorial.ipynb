```python
def factorial(n):
  if n == 0:
    return 1;
  else:
    for i in range(1, n):
      n *= i;
  return n;
```

```python
factorial(5)
```

```python
def recursive_factorial(n):
  if n == 0:
    return 1;
  return n * recursive_factorial( n - 1);
```

```python
recursive_factorial(5)
```

```python
def ka_fani_factorial(n):
  result =1;
  for i in range(n):
    result *= (n - i);
  return result;
```
